# Checking correlation

```python
import seaborn as sns

def corr_plot(df):
    dataplot = sns.heatmap(df.corr(), cmap="YlGnBu", annot=True)
    return dataplot
```

```python
corr_plot_dat = corr_plot(df[df.columns.difference(['Country'])])
plt.title('Correlation Plot')
plt.gcf().autofmt_xdate()
plt.show()
```

```python
df_correlated = df[df.columns.difference(['Country'])].corr()
df_correlated
```

# Exploring variables

```python
half_q3_Global_innovation_indices = (df_basic_stats.Global_innovation_indices.loc['75%'] + df_basic_stats.Global_innovation_indices.loc['max'])/2
half_q3_Business_Sophistication = (df_basic_stats.Business_Sophistication.loc['75%'] + df_basic_stats.Business_Sophistication.loc['max'])/2
```

```python
half_q1_Global_innovation_indices = (df_basic_stats.Global_innovation_indices.loc['25%'] + df_basic_stats.Global_innovation_indices.loc['min'])/2
half_q1_Business_Sophistication = (df_basic_stats.Business_Sophistication.loc['25%'] + df_basic_stats.Business_Sophistication.loc['min'])/2
```

```python
df_advanced = df.loc[(df.Global_innovation_indices > df_basic_stats.Global_innovation_indices.loc['75%']) 
                     & 
                     (df_basic_stats.Business_Sophistication.loc['75%'] < df.Business_Sophistication), ['Country']]
df_advanced
```

```python
df_contaminant = df.loc[df.Pollution > 73.885000, ['Country']]
df_contaminant
```

```python
int_df = pd.merge(df_advanced, df_contaminant, how ='inner', on =['Country', 'Country'])
print(int_df)
```

```python
df_advanced = df.loc[(df.Global_innovation_indices < half_q1_Global_innovation_indices) 
                     & 
                     (half_q3_Business_Sophistication < df.Business_Sophistication), ['Country']]
df_advanced
```

```python
df_worst = df.loc[(df.Global_innovation_indices < half_q1_Global_innovation_indices) 
                     & 
                     (half_q1_Business_Sophistication > df.Business_Sophistication), ['Country']]
df_worst
```
